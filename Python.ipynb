```python
def calcula_idade_maior(idade_monica, idade_filho1, idade_filho2):
    # Calcula a idade do terceiro filho
    idade_filho3 = idade_monica - idade_filho1 - idade_filho2

    # Utiliza a função max para calcular a maior idade entre os 3 filhos
    idade_maxima = max(idade_filho1, idade_filho2, idade_filho3)

    # Retorna a idade do terceiro filho e a idade do filho mais velho
    return idade_filho3, idade_maxima


# Chamada da função e exibição do resultado
idade_filho3, idade_mais_velho = calcula_idade_maior(68, 12, 30)
print(f"Idade do terceiro filho: {idade_filho3}")
print(f"Idade do filho mais velho: {idade_mais_velho}")
```

```python
def maior_livro(paginas):
    # A função recebe uma lista de números e retorna o maior deles
    return max(paginas)


# Programa principal
print("Descubra qual livro tem mais páginas!")

# Coleta dos dados
livros = []
for i in range(1, 6):
    num_paginas = int(input(f"Digite a quantidade de páginas do livro {i}: "))
    livros.append(num_paginas)

# Chama a função e mostra o resultado
maior = maior_livro(livros)
print(f"\nO livro com mais páginas tem {maior} páginas.")
```
